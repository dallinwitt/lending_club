```python
loans.groupby('loan_status')['loan_status'].count()
```

```python
#we'll only look at loans that are either charged off or fully paid
#change 'Does note meet...' to their shorter name
status_dict = {'Does not meet the credit policy. Status:Charged Off':'Charged Off',
           'Does not meet the credit policy. Status:Fully Paid':'Fully Paid'}

loans['loan_status'] = loans['loan_status'].replace(status_dict)

loans = loans[(loans['loan_status'] == 'Fully Paid') | (loans['loan_status'] == 'Charged Off')]
```

```python
#check to make sure that the df is what we want
loans.groupby('loan_status')['loan_status'].count()
```

```python
#create a dict to convert the employment lengths to ints
empl_dict = {'10+ years':10, 
             '6 years':6, 
             '4 years':4, 
             '< 1 year':0.5, 
             '2 years':2,
             '9 years':9, 
             '5 years':5, 
             '3 years':3, 
             '7 years':7, 
             '1 year':1,
             '8 years':8,
            np.nan:0}

loans['emp_length'] = loans['emp_length'].replace(empl_dict).astype(int, errors = 'ignore')
```
